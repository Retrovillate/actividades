```python
import math
class Numero:
    
    normal = 1
    romano ="I"
    
    # metodo que transforma a un entero a romano
        
    def integerToRoman(self,numero):
        romansDict = \
            {
                1: "I",
                5: "V",
                10: "X",
                50: "L",
                100: "C",
                500: "D",
                1000: "M",
                5000: "G",
                10000: "H"
            }
        
        div = 1
        while numero >= div:
            div *= 10
        div /= 10
        
        res = ""
        
        while numero:
            
            lastNum = int(numero / div)
            
            if lastNum <= 3:
                res += (romansDict[div] * lastNum)
            elif lastNum == 4:
                res += (romansDict[div] + romansDict[div * 5])
            elif 5 <= lastNum <= 8:
                res += (romansDict[div * 5] + (romansDict[div] * (lastNum -5)))
            elif lastNum == 9:
                res += (romansDict[div] + romansDict[div * 10])
            
            numero = math.floor(numero % div)
            div /= 10
                                                     
        return res
    
    # metodo que tranforma romano a entero
    
    def roman2int(self, romval):
        romanos = {'I': 1, 'V': 5, 'X': 10,'L': 50, 'C': 100, 'D': 500, 'M': 1000, 'G': 5000, 'H': 10000}
        total = 0
        prev = 0
        for letra in romval[::-1]:
            valor = romanos[letra]
            total += valor if valor >= prev else -valor
            prev = valor

        return total
        
    
    # constructor
    
    def __init__(self, numero):
        self.normal = numero
        self.romano = self.integerToRoman(numero)
        
    # imprime
    
    def imprime(self):
        print(str(self.normal) + " " + self.romano)
        
    # Suma romano
    
    def suma_romano(self, romanoSum):
        
        esRomano = self.is_romano(romanoSum)
        
        if (esRomano == True):
            numeroOrig = self.normal
            numeroSumar = self.roman2int(romanoSum)
            SumaFinal = numeroOrig + numeroSumar
            self.normal = SumaFinal
            self.romano = self.integerToRoman(SumaFinal)
            
        else:
            print('Ha fallado el número', romanoSum)   
        
    # Validar si es un romano
    
    def is_romano(self, romano):
        romanos = ['I', 'V', 'X','L', 'C', 'D', 'M', 'G', 'H']
        salida = False
        posMax = -2
        cont = 0
        contSeg = 0;
        anterioRomano = "K"
        maxSeguidoRoman = 3
        minValorRomanoNoRep = 1
        
        try:
            tamaStr = len(romano)
            tamaIni = 0
            tamaUlt = tamaStr-1
            pasos = -1
            finCiclo = -1
            for i in range(tamaUlt, finCiclo, pasos):
                
                actualRomano = romano[i]
                
                pos = romanos.index(romano[i])
            
                if(i >= tamaIni):
                    if(posMax <= pos) :
                        cont += 1
                        posMax = pos
                        
                        if(actualRomano == anterioRomano):
                            
                            contSeg+=1
                        else:
                            contSeg = 1
                            anterioRomano = actualRomano
                        
                        esPar = False
                        if(pos % 2 == 0):
                            esPar = True
                            
                            if(contSeg > maxSeguidoRoman):
                                
                                return salida
                        else:
                            
                            if(contSeg > minValorRomanoNoRep):
                                return salida              
                            
                    elif((i+1) <= tamaUlt):
                        posAnt = romanos.index(romano[i+1])
                        
                        if(esPar == True):
                            
                            posRes = posAnt -2
                            
                            if(posRes == pos):
                                
                                cont += 1  
                            else:
                                return salida
                        else:
                            posRes = posAnt -1
                            
                            if(posRes == pos):
                                
                                cont += 1  
                            else:
                                return salida
                    else:
                        return salida
                else:
                    print("no va llegar a este punto")
                    
             # evaluar si todos los caracteres son romanos
            if(cont == tamaStr):
                salida=True
                
                     
        except:
            return salida
        return salida   
        
```
